```python
#функции выводящие графики
def countplot(x, hue, data, ex: str):
    ax = sns.countplot(x=x, hue=hue, data=data)
    ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)
    plt.title(ex)
    plt.show()

def box(x, y, data, ex: str):
    sns.boxplot(x=x, y=y, data=data)
    plt.title(ex)
    plt.show()

```

```python
def base_metric_count(ex: str, df, column):
  #функция с изучением числовых столбцов
    print('Base metrics')
    print(df[column].describe())
    print('*' * 80)

    print('Must popular value of metric, top 5')
    print(df[column].value_counts().nlargest(5))
    print('*' * 80)

    print('Excess', kurtosis(df[column]))
    print('Assymetry', skew(df[column]))

    sns.distplot(df[column])
    plt.title(ex + '  ' + column)
    plt.show()
```

```python
def base_categ(ex: str, df, column, plot=countplot, y=None):
  #фун-я с изучением категориальных столбцов
    print('Base metrics')
    print(df[column].describe())
    print('*' * 80)

    print('Ratio value')
    print(df[column].value_counts(normalize=True))
    print('*' * 80)

   

    plot(column, y, df, ex + '  ' +column)
```

```python
repl = {'Yes': 1, 'No': 0, 'No internet service': 0, 'No phone service': 0,\
        'Male': 1, 'Female': 0,\
 'Fiber optic': 1, 'DSL': 2, 'Two year': 2, 'Month-to-month': 0, 'One year': 1,\
 'Electronic check': 0, 'Bank transfer (automatic)': 1,\
  'Mailed check': 0, 'Credit card (automatic)': 1}
#словарь для замены строк на цифры

def to_one_type(exp: dict, data):
  #заменяем по словарю, обращаем в число
    data = data.replace(exp).astype(float, errors = 'ignore')
    return data

def decorate_column(data):
  #создаем новые столбцы
    df_trans = copy.deepcopy(data)#копируем дф
    df_trans = to_one_type(repl, df_trans)#заменеяем на цифры

    df_trans['socialLoad'] = df_trans['Partner'] + df_trans['SeniorCitizen'] + df_trans['Dependents']
    df_trans['supportLoad'] = df_trans['DeviceProtection'] + df_trans['TechSupport'] 
    df_trans['streamLoad'] = df_trans['StreamingTV'] + df_trans['StreamingMovies'] 
    df_trans['onlineLoad'] = df_trans['OnlineSecurity'] + df_trans['OnlineBackup'] 
    df_trans['loyality'] =  df_trans['Contract'].apply(lambda x: 0 if x == 'Month-to-month' else 1)
     #складываем числовые показатели
    df_trans.drop(['Partner', 'SeniorCitizen', 'Dependents', 'DeviceProtection', 'TechSupport',\
               'StreamingTV', 'StreamingMovies', 'OnlineSecurity', 'OnlineBackup'], axis=1,\
              inplace=True)#отбрасываем столбцы, из которых делали новые столбцы

    return df_trans
    
df_trans = decorate_column(to_one_type(repl, df))#создаем дф с новыми столбцами
```

```python
def get_bootstrap_samples(data, n_samples):
  #функция бустрапирования
    indices = np.random.randint(0, len(data), (n_samples, len(data)))
    #пределяем начало и конец отчёта, кол-во осей и кол-во эл-ов на каждой оси
    samples = data[indices]
    #набираем значений из дф по созданным индексам
    return samples
  
```

```python
def stat_intervals(stat, alpha):
  #ф-я доверительного интервала
    boundaries = np.percentile(stat, [100 * alpha / 2., 100 * (1 - alpha / 2.)])
    #вычиляем границы, даем данные и указываем какие персентили нужны
    return boundaries
```

Выдвигаем нулевую гипотезу, что функция распределения у выборок равна. Проверяем перестановочным критерием.

```python
def permutation_t_stat_ind(sample1, sample2):
  #вычисляем t-статистику
    return np.mean(sample1) - np.mean(sample2)
```

```python
def get_random_combinations(n1, n2, max_combinations):
   #перемешиваем данные
    index = list(range(n1 + n2)) #совмещаем дины обеих выборок
    indices = set([tuple(index)])#оставляем только не повторяющиеся
    for i in range(1, max_combinations):
      np.random.shuffle(index)#перемешиваем
      indices.add(tuple(index))#добавляем
    return [(index[:n1], index[n1:]) for index in indices]
```

```python
def permutation_zero_dist_ind(sample1, sample2, max_combinations=None):
  #вычиляем нулевое распределение
 
    joined_sample = np.hstack((sample1, sample2))#cоединяем массивы в один одноразмерный
    n1 = len(sample1)
    n = len(joined_sample)
  #узнаем длины массивов
    if max_combinations:
        indices = get_random_combinations(n1, len(sample2), max_combinations)
        #генерируем перестановку
    else:
        indices = [(list(index), filter(lambda i: i not in index, range(n))) \
               for index in itertools.combinations(range(n), n1)]
#генерируем список без повторений, делаем кортеж из индекса, создаем лист,где нет э-в из индекаса 
               
    distr = [joined_sample[list(i[0])].mean() - joined_sample[list(i[1])].mean() \
             for i in indices]
             #ура нулевое распределние
    return distr
```

```python
def permutation_test(sample, mean, max_permutations=None, alternative='two-sided'):
    if alternative not in ('two-sided', 'less', 'greater'):
        raise ValueError('alternative not recognized\n'
        'should be "two-sided", "less", "greater"')
    #условие кртитических областей    

    t_stat = permutation_t_stat_ind(sample, mean)

    zero_distr = permutation_zero_dist_ind(sample, mean, max_permutations)
#вычисляем вероятность получения статистики как в эксперементе при справедливоасти
#нулевой гипотезы
    if alternative == 'two-sided':
        return sum([1. if abs(x) >= abs(t_stat) else 0. for x in zero_distr]) / len(zero_distr)

    if alternative == 'less':
         return sum([1. if x <= t_stat else 0. for x in zero_distr]) / len(zero_distr)

    if alternative == 'greater':
         return sum([1. if x >= t_stat else 0. for x in zero_distr]) / len(zero_distr)
        
```

Ставим нулеву гипотезу, что отношение женатых неженатых людей в выборках одинаковое.  Проверяем с помощь Z критерия для зависимых выборок.

```python
from scipy import stats
import scipy

def proportions_diff_z_stat_ind(sample1, sample2):
#двупропорциональный z-test
    sample = list(zip(sample1, sample2)) 
    n = len(sample)
   
    f = sum([1 if (x[0] == 1 and x[1] == 0) else 0 for x in sample])
    g = sum([1 if (x[0] == 0 and x[1] == 1) else 0 for x in sample])
  #таблица сопряжённости

    return float(f - g) / np.sqrt(f + g - float((f - g)**2) / n )
     #
```

```python
def proportions_diff_z_test(z_stat, alternative = 'two-sided'):
    if alternative not in ('two-sided', 'less', 'greater'):
        raise ValueError("alternative not recognized\n"
                         "should be 'two-sided', 'less' or 'greater'")
       #условие кртитических областей    
    if alternative == 'two-sided':
        return 2 * (1 - scipy.stats.norm.cdf(np.abs(z_stat)))

    if alternative == 'less':
        return scipy.stats.norm.cdf(z_stat)

    if alternative == 'greater':
        return 1 - scipy.stats.norm.cdf(z_stat)
        #создаем неприрывные случайные величины с кумултяивными функциями распределения
```

```python
class Stacking:

    def __init__(self, meta_alg, data, targets, metrics, models, random_state=None):
       #указываем обязательные параметры
        self.models = models
        self.meta_alg = meta_alg
        self.data = data
        self.targets = targets
        self.metrics = metrics
        
        self.random_state = random_state
        

    def __call__(self, p=0.3, cv=5, parameters=None, meta_params=None, test_size=0.3, test=None, targets_test=None):
        #для вызова с условием есть или нет тестовая часть
        self.test = test
        self.targets_test = targets_test

        if self.test is None:
            self.pipline(test_size=test_size)
            self.learn_models(p=p, cv=cv,parameters=parameters, meta_params=meta_params)
            self.predict(self.data_test)
            return self.error_score()
        else:
            self.learn_models(X=self.data, y=self.targets, p=p, cv=cv,\
                              parameters=parameters, meta_params=meta_params)
            self.predict(self.test)
            
            return self.error_score(self.targets_test)

  

    def pipline(self, test_size=0.3):
       #делим данные на обучающие и тестовые
        self.data_train, self.data_test, self.targets_train, self.targets_test = train_test_split(self.data, self.targets,
                                                              test_size=test_size, random_state=self.random_state,  stratify=self.targets)
    
        return self.data_train, self.data_test, self.targets_train, self.targets_test   

    def meta_grid(self, y):
       #применяем поиск лучших пара-в к метаалгоритму, если етсь
        self.meta_alg = self.meta_alg(random_state=self.random_state) if hasattr(self.meta_alg(), 'random_state') else self.meta_alg()

        if self.meta_params is None:
            self.meta_alg.fit(self.valid, y)
        else:
            gcv = GridSearchCV(self.meta_alg, self.meta_params, cv=self.skf)
            gcv.fit(self.valid, y)
            meta_alg = gcv.best_estimator_   
            print(meta_alg)           
            self.meta_alg = meta_alg.fit(self.valid, y)
    
    def learn_models(self, X=None, y=None, p=0.2, cv=5, parameters=None, meta_params=None):
         #обучаем модели
        self.p = p 
        self.cv = cv
        self.parameters = parameters
        self.meta_params = meta_params
        #делим на куски с одинаковым соотношением классов
        self.skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=self.random_state)
        model_list = [model(random_state=self.random_state) if hasattr(model(), 'random_state') else model() for model in self.models]
        models = []
 
        if (X is None) & (y is None):
            X = self.data_train
            y = self.targets_train
      
        if (self.p > 0): 
            #разбиваем на обучение и тест
            train, valid, y_train, y_valid = train_test_split(X, y, test_size=self.p, random_state=self.random_state)
            #создаем пустую матрицу для предсказаний
            self.valid = np.empty((valid.shape[0], len(self.models)))
           
            for n, model in enumerate(model_list):
               #проверяем есть ли пар-ры выбираем лучшие для моделй
                if self.parameters is None:
                    model.fit(train, y_train)
                    prediction= model.predict(valid)
                    self.valid[:, n] = prediction
                    print(self.metrics(y_valid, prediction))
                    
                else:
                    for i, parameter in enumerate(self.parameters):
                       
                        if i == n:
                            gcv = GridSearchCV(model, self.parameters[i], cv=self.skf)
                            gcv.fit(train, y_train)
                            model = gcv.best_estimator_ 
                            print(model)
                            model.fit(train, y_train)
                            prediction = model.predict(valid)
                            self.valid[:, n] = prediction
                            models.append(model)
                            print(self.metrics(y_valid, prediction))
            #сохраняем модели и метамодель    
            self.models = models
            self.meta_grid(y_valid)

        else: 
    #создаем пустую матрицу для предсказаний
            self.valid = np.empty(X.shape[0], len(self.models))
            for n, model in enumerate(self.models):
                #проверяем есть ли пар-ры выбираем лучшие для моделй
                if self.parameters is None:
                    prediction = cross_val_predict(model, X, y, cv=self.cv, n_jobs=-1, method='predict')
                    self.valid[:, n] = prediction
                    model.fit(X, y)
                    print(self.metrics(y, prediction))
                    
                else:
                    for i, parameter in enumerate(self.parameters):
                        if i == n:
                            gcv = GridSearchCV(model, self.parameters[i], cv=self.skf)
                            gcv.fit(X, y)
                            model = gcv.best_estimator_ 
                            print(model)
                            prediction = cross_val_predict(model, X, y, cv=self.cv, n_jobs=-1, method='predict')
                            self.valid[:, n] = prediction
                            model.fit(X, y)
                            print(self.metrics(y, prediction))
    #сохраняем модели и метамодель                         
            self.models = models
            self.meta_grid(y)

        return self.models
                    
    def predict(self, data_predict):
      #создаем матрицу для предсказаний
        X_meta = np.empty((data_predict.shape[0], len(self.models)))
#предсказываем по моделям получившимся
        for n, model in enumerate(self.models):
            X_meta[:, n] = model.predict(data_predict)
        #предсказываемпо метамодели 
        self.prediction =  self.meta_alg.predict(X_meta)

        
        return self.prediction

    
    def error_score(self, targets_test):
      #получаем метрику
        print(f'Stacking score:{self.metrics(targets_test, self.prediction)}')
        return self
        
        
```

```python
def calibrated(model, X, y):
  #фун-я калибровки модели
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=17)
    #делим на отрезки с таким же соотношением данных как в выборке
    calibrator = CalibratedClassifierCV(model, cv=skf)
    calibrator.fit(X, y)
    #калибруем
    return calibrator
```

```python
def obj(features_upsampled, target_upsampled, \
        valid, target_valid, model, calibrated=False):
    if calibrated == False:
        model.fit(features_upsampled, target_upsampled)
        preds = model.predict(valid)
    #обучаем модель
    else:
        calibrator = calibrated(model, features_upsampled, target_upsampled,)
        preds = calibrator.predict(valid)
 #предсказываем
    auc = roc_auc_score(target_valid, preds, average ='micro')
    f1 = f1_score(target_valid, preds, average='micro')
    print_all_metrics(target_valid, preds, str(model)) #выводим метрики
    return f1, auc, preds, model
```

```python
auc = []
pred = []
models = []
f1 = []

def all_type_features_model(features_upsampled, target_upsampled, \
        X_valid, y_valid, X_train_st, X_test_st, X_clust, X_clust_test, model):
  #функция в которой  мы обучаем модели и сохраняем их показатели в листы
    f1_0, auc_0, preds_0, model_0 = obj(features_upsampled, target_upsampled, \
        X_valid, y_valid, model)
    f1.append(f1_0)
    auc.append(auc_0)
    pred.append(preds_0)
    models.append((model_0, 'with upsampled'))

    f1_1, auc_1, preds_1, model_1 = obj(X_train_st, target_upsampled, \
        X_test_st, y_valid, model)
    f1.append(f1_1)
    auc.append(auc_1)
    pred.append(preds_1)
    models.append((model_1, 'with StandScal and upsampled'))

    f1_2, auc_2, preds_2, model_2 = obj(X_clust, y_train, \
        X_clust_test, y_valid, model)
    f1.append(f1_2)
    auc.append(auc_2)
    pred.append(preds_2)
    models.append((model_2, 'with Calibrated, column clust and StratFold'))
    
    
    return max(auc), max(f1)
```
